```python
# Import necessary libraries
import cv2
from deepface import DeepFace
import matplotlib.pyplot as plt

# Load the image
image_path = "he.jpg"  # Replace with your image path
img = cv2.imread(image_path)

# Convert image from BGR (OpenCV default) to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display the input image
plt.imshow(img_rgb)
plt.axis('off')
plt.title("Input Image")
plt.show()

try:
    # Analyze the emotions in the image
    analysis = DeepFace.analyze(img_path=image_path, actions=['emotion'])

    # Extract emotions and confidence scores
    dominant_emotion = analysis[0]['dominant_emotion']
    emotions = analysis[0]['emotion']

    # Print results
    print(f"Detected Emotion: {dominant_emotion}")
    print("Emotion Confidence Scores:")
    for key, value in emotions.items():
        print(f"{key.capitalize()}: {value:.2f}%")

    # Plot emotion confidence scores
    plt.figure(figsize=(8, 4))
    plt.bar(emotions.keys(), emotions.values(), color='skyblue')
    plt.xlabel("Emotions")
    plt.ylabel("Confidence (%)")
    plt.title("Emotion Analysis")
    plt.show()

except Exception as e:
    print("Error:", e)
```
